* **Valeur la plus fréquente**: Valeur qui apparaît le plus souvent.
* **Fréquence de la valeur la plus fréquente**: Nombre de fois que la valeur la plus fréquente apparaît.

```python
# Colonnes où nous voulons supprimer les lignes avec des valeurs NaN
columns_to_check = [
    'id',
    'num_voie',
    'code_postal',
    'id_division',
    'ida3d_enb',
    'id_eqpt',
    'competence',
    'url_plan'
]

# Supprimer les lignes où ces colonnes contiennent des valeurs NaN
data_cleaned_final = data_cleaned.dropna(subset=columns_to_check)

# Afficher les statistiques après suppression
print("Nombre de lignes après suppression :")
print(data_cleaned_final.shape[0])

print("Nombre de colomnes après suppression :")
print(data_cleaned_final.shape[1])

print("\nNombre de valeurs manquantes après suppression :")
print(data_cleaned_final.isna().sum())
```


on va remplacer Nan avec 'inconnu'
type_voie                   247
libelle_voie                247

```python
# Remplacer les valeurs NaN par 'inconnu' dans les colonnes spécifiques
columns_to_replace = ['type_voie', 'libelle_voie']

data_cleaned_final = data_cleaned.copy()
data_cleaned_final[columns_to_replace] = data_cleaned_final[columns_to_replace].fillna('inconnu')

# Afficher les statistiques après remplacement
print("Nombre de valeurs manquantes après remplacement :")
print(data_cleaned_final[columns_to_replace].isna().sum())


print("\nNombre de valeurs manquantes après suppression :")
print(data_cleaned_final.isna().sum())
```

On va appliquer le median pour les ligne contient Nan values de cetts colomnes
surface_calculee            242
superficie_totale_reelle    487
surface_horticole           542
circonference_cm            493


```python
# List of numerical columns with missing values
numerical_columns = [
    'surface_calculee',
    'superficie_totale_reelle',
    'surface_horticole',
    'circonference_cm'
]

# Replace NaN with the median for each column
for col in numerical_columns:
    median_value = data_cleaned_final[col].median()
    data_cleaned_final[col].fillna(median_value, inplace=True)
```

# Size Analysis

#### Surface Area Distribution: 

Surface Area Distribution: Compare the surface_calculee and superficie_totale_reelle columns to see discrepancies in reported vs. actual sizes of green spaces.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate the sum of surface areas by category
surface_calculee_by_category = data_cleaned_final.groupby("categorie")["surface_calculee"].sum()
superficie_totale_reelle_by_category = data_cleaned_final.groupby("categorie")["superficie_totale_reelle"].sum()

# Create a DataFrame for comparison
comparison_df = pd.DataFrame({
    "Surface Calculée": surface_calculee_by_category,
    "Superficie Totale Réelle": superficie_totale_reelle_by_category
})

# Plot side-by-side bar charts
fig, ax = plt.subplots(figsize=(14, 8))
comparison_df.plot(kind='bar', ax=ax)
plt.title("Comparison of Calculated Surface vs Actual Surface by Category", fontsize=16)
plt.xlabel("Category", fontsize=14)
plt.ylabel("Surface Area", fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Surface Type', title_fontsize='13', fontsize='12')

# Display the plot
plt.tight_layout()
plt.show()

# Print the DataFrame for reference
print("Surface Area Comparison by Category:")
print(comparison_df)
```

#### Category-wise Size Analysis:

Category-wise Size Analysis: Analyze how the size of green spaces varies by category, providing insight into the scale of different types of green spaces.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group by 'categorie' and calculate total and average sizes for both surface measurements
category_stats = data_cleaned_final.groupby('categorie').agg({
    'surface_calculee': ['sum', 'mean'],
    'superficie_totale_reelle': ['sum', 'mean']
}).reset_index()

# Rename columns for easier reference
category_stats.columns = [
    'Categorie', 
    'Total Surface Calculée', 'Average Surface Calculée', 
    'Total Superficie Totale Réelle', 'Average Superficie Totale Réelle'
]

# Plot total sizes by category
fig, ax = plt.subplots(figsize=(14, 8))

category_stats.plot(kind='bar', x='Categorie', y=['Total Surface Calculée', 'Total Superficie Totale Réelle'], ax=ax, color=['#1f77b4', '#ff7f0e'])
plt.title('Total Size of Green Spaces by Category', fontsize=16)
plt.xlabel('Category', fontsize=14)
plt.ylabel('Total Size', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Surface Type', title_fontsize='13', fontsize='12')

# Display the plot
plt.tight_layout()
plt.show()

# Plot average sizes by category
fig, ax = plt.subplots(figsize=(14, 8))

category_stats.plot(kind='bar', x='Categorie', y=['Average Surface Calculée', 'Average Superficie Totale Réelle'], ax=ax, color=['#1f77b4', '#ff7f0e'])
plt.title('Average Size of Green Spaces by Category', fontsize=16)
plt.xlabel('Category', fontsize=14)
plt.ylabel('Average Size', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Surface Type', title_fontsize='13', fontsize='12')

# Display the plot
plt
```

```python
import matplotlib.pyplot as plt

# Calculer la somme des surfaces par catégorie pour chaque colonne
surface_calculee_par_categorie = data_cleaned_final.groupby("categorie")["surface_calculee"].sum()
superficie_totale_reelle_par_categorie = data_cleaned_final.groupby("categorie")["superficie_totale_reelle"].sum()

# Créer un DataFrame pour faciliter la comparaison
comparison_df = pd.DataFrame({
    "Surface Calculée": surface_calculee_par_categorie,
    "Superficie Totale Réelle": superficie_totale_reelle_par_categorie
})

# Calculer les proportions pour chaque catégorie
comparison_df["Surface Calculée (%)"] = (comparison_df["Surface Calculée"] / comparison_df.sum(axis=1)) * 100
comparison_df["Superficie Totale Réelle (%)"] = (comparison_df["Superficie Totale Réelle"] / comparison_df.sum(axis=1)) * 100

# Créer un diagramme de progression (100% stacked bar chart) avec une plus grande taille
comparison_df[["Surface Calculée (%)", "Superficie Totale Réelle (%)"]].plot(kind='barh', stacked=True, figsize=(20, 12), color=['#1f77b4', '#ff7f0e'])
plt.xlabel("Pourcentage (%)")
plt.ylabel("Catégorie")
plt.title("Comparaison des Surfaces calculée par Catégorie (Diagramme de Progression)")
plt.legend(loc='lower right')
plt.show()
```

```python
# Importer les bibliothèques nécessaires
from shapely.geometry import Polygon
import matplotlib.pyplot as plt
import json

# Récupérer la première ligne de la colonne 'geo_shape'
first_row_geo_shape = data_cleaned_final['geo_shape'].iloc[16]

# Charger les données JSON pour extraire les coordonnées
geo_shape_data = json.loads(first_row_geo_shape)
coordinates = geo_shape_data['coordinates'][0]

# Créer un objet Polygon avec les coordonnées extraites
polygon = Polygon(coordinates)

# Extraire les coordonnées pour la visualisation
x, y = polygon.exterior.xy

# Visualiser le polygone
plt.figure(figsize=(8, 6))
plt.fill(x, y, alpha=0.5, fc='blue', edgecolor='black')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Visualisation du Polygone')
plt.grid(True)
plt.show()
```

# Correlation Analysis 

To investigate potential correlations between different variables, you can create a heatmap of correlations and scatter plots to explore relationships between variables such as size, category, and the presence of features.

### Heatmap of Correlations

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming data_cleaned_final is already loaded and relevant columns are numeric

# Compute correlation matrix
corr_matrix = data_cleaned_final[['surface_calculee', 'superficie_totale_reelle', 'circonference_cm']].corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heatmap des Corrélations entre Variables')
plt.show()
```

### Scatter Plot: Surface Size vs. Circonference

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming data_cleaned_final is already loaded

# Scatter plot for surface_calculee vs. circonference_cm
plt.figure(figsize=(12, 8))
plt.scatter(data_cleaned_final['surface_calculee'], data_cleaned_final['circonference_cm'], alpha=0.5)
plt.xlabel('Surface Calculée')
plt.ylabel('Circonférence en cm')
plt.title('Relation entre Surface Calculée et Circonférence')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming data_cleaned_final is already loaded

# Create a regression plot for surface_calculee vs. circonference_cm
plt.figure(figsize=(12, 8))
sns.regplot(x='surface_calculee', y='circonference_cm', data=data_cleaned_final, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.xlabel('Surface Calculée')
plt.ylabel('Circonférence en cm')
plt.title('Relation entre Surface Calculée et Circonférence avec Régression Linéaire')
plt.grid(True)
plt.show()
```
